```python
# without queue
tree = {
    "a": ["b", "c"],
    "b": ["d", "e"],
    "c": ["f"],
    "d": [],
    "e": [],
    "f": ["g"],
    "g": []
}

start= ["a"] 
goal = "g"
visited = []

while start:
    a = start.pop(0)   
    if a not in visited:
        visited.append(a)
        if a == goal:
            print("Your goal met =", visited)
            break
        for i in tree[a]:
            start.append(i)
```

```python
# with queue
class BFS:
    def __init__(self,tree):
        self.tree=tree
        self.visited=[]
        self.queue=[]
    def search(self, start, goal):
        self.queue.append(start)

        while self.queue:
            a = self.queue.pop(0)   
            if a not in self.visited:
                self.visited.append(a)
                if a == goal:
                    print("Your goal met =", self.visited)
                    break
                for i in self.tree[a]:
                    self.queue.append(i)
tree = {
    "a": ["b", "c"],
    "b": ["d", "e"],
    "c": ["f"],
    "d": [],
    "e": [],
    "f": ["g"],
    "g": []
}
start = "a"
goal = "g"
b = BFS(tree)
b.search(start, goal)
```
