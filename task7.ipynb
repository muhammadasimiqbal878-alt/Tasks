# without pirorty queue

```python
class Graph:
    def __init__(self, adjacency_list):
        self.adjacency_list = adjacency_list
    def get_neighbors(self, v):
        return self.adjacency_list.get(v)
    def h(self, n):
        H = {
            'A': 1,
            'B': 1,
            'C': 1,
            'D': 1
        }
        
        return H.get(n) 
    def a_star_algorithm(self, start_node, stop_node):
       
        if start_node not in self.adjacency_list or stop_node not in self.adjacency_list:
            print("Invalid start or goal node! Please enter correct node names.")
            return None

        open_list = set([start_node])
        closed_list = set([])

        g = {}
        g[start_node] = 0

        parents = {}
        parents[start_node] = start_node
        while len(open_list) > 0:
            n = None
            for v in open_list:
                if n is None or g[v] + self.h(v) < g[n] + self.h(n):
                    n = v

            if n is None:
                print(" Path does not exist!")
                return None

            if n == stop_node:
                reconst_path = []
                while parents[n] != n:
                    reconst_path.append(n)
                    n = parents[n]
                reconst_path.append(start_node)
                reconst_path.reverse()
                print("Path found:", reconst_path)
                return reconst_path

            for (m, weight) in self.get_neighbors(n):
                if m not in open_list and m not in closed_list:
                    open_list.add(m)
                    parents[m] = n
                    g[m] = g[n] + weight
                else:
                    if g[m] > g[n] + weight:
                        g[m] = g[n] + weight
                        parents[m] = n
                        if m in closed_list:
                            closed_list.remove(m)
                            open_list.add(m)

            open_list.remove(n)
            closed_list.add(n)

        print("Path does not exist!")
        return None
adjacency_list = { 
    'A': [('B', 1), ('C', 3), ('D', 7)], 
    'B': [('D', 5)], 
    'C': [('D', 12)],
    "D":[]
    } 

graph1 = Graph(adjacency_list)
graph1.a_star_algorithm('A','D')
graph1.a_star_algorithm('A', 'F')

```

# with pirority queue

```python
import heapq

class Graph:
    def __init__(self, adjacency_list):
        self.adjacency_list = adjacency_list

    def get_neighbors(self, v):
        return self.adjacency_list.get(v, [])

    def h(self, n):
       
        H = {
            'A': 1,
            'B': 1,
            'C': 1,
            'D': 1
        }
        return H.get(n, float('inf'))

    def a_star_priority(self, start_node, stop_node):

        if start_node not in self.adjacency_list or stop_node not in self.adjacency_list:
            print("Invalid start or goal node! Please enter correct node names.")
            return None

        
        open_heap = []
      
        g = {start_node: 0}
     
        parents = {start_node: start_node}

        heapq.heappush(open_heap, (g[start_node] + self.h(start_node), start_node))

        closed_set = set()

        while open_heap:
           
            current_f, n = heapq.heappop(open_heap)

         
            if n in closed_set:
                continue

          
            if n == stop_node:
                path = []
                while parents[n] != n:
                    path.append(n)
                    n = parents[n]
                path.append(start_node)
                path.reverse()
                print("Path found:", path)
                return path

            
            closed_set.add(n)

       
            for (m, weight) in self.get_neighbors(n):
                if m in closed_set:
                    continue

                tentative_g = g[n] + weight

             
                if m not in g or tentative_g < g[m]:
                    g[m] = tentative_g
                    parents[m] = n
                    f_m = tentative_g + self.h(m)
                   
                    heapq.heappush(open_heap, (f_m, m))

   
        print("Path does not exist!")
        return None


adjacency_list = {
    'A': [('B', 1), ('C', 3), ('D', 7)],
    'B': [('D', 5)],
    'C': [('D', 12)],
    'D': []
}

graph1 = Graph(adjacency_list)
graph1.a_star_priority('A', 'D')   
graph1.a_star_priority('A', 'F')  
```
